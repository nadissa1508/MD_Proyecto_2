```python
df = df.unique()
```

```python
duplicate_code_numbers = df.group_by(
    pl.col("Sample code number")
).len().filter(
    pl.col('len') > 1
).select(
    pl.col('Sample code number')
).to_series()

duplicate_code_numbers.sort()
```

```python
df_duplicate_by_code_number = df.filter(
    pl.col('Sample code number').is_in(duplicate_code_numbers.implode()) 
).sort(pl.col('Sample code number')).head( 98)
df_duplicate_by_code_number
```
